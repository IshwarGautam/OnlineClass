```python
import matplotlib
print(matplotlib.__version__)
print(matplotlib.get_backend())
```

```python
matplotlib.use('nbagg') #The nbagg backend
#"nbagg" enables interactive figures in a live IPython notebook session
matplotlib.get_backend()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(facecolor=(1,0,1,0.5)) #RGBA value (0<=1) where A is alpha channel, specify the opacity of color
plt.show()
```

```python
plt.figure(figsize=plt.figaspect(0.1),facecolor=(1,0,1,0.5))
plt.show()
```

```python
fig = plt.figure()
splot = plt.subplot(111) #1 <= num <= 1
splot.set(xlim=[0.5, 3], ylim=[-1, 1], title="Example of axes",
      xlabel='X-axis', ylabel='Y-axis')
plt.show()
```

```python
fig = plt.figure()
splot = plt.subplot(111)
splot.set_xlim([0.5, 3])
splot.set_ylim([-1, 1])
splot.set_title('Second Example of axes')
splot.set_ylabel('Y-Axis', size=20)
splot.set_xlabel('X-Axis', size=20)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot([1,2,3,4],[10,20,30,40], color='green', linewidth=5)
ax.scatter([0.5,1.5,2.5,5], [5,10,15,20], c=[4,3,2,1], marker='*')
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
axes[0,0].set(title='Top Left')
axes[0,1].set(title='Top Right')
axes[1,0].set(title='Buttom Left')
axes[1,1].set(title='Buttom Right')

# To iterate over all items in a multidimensional numpy array, we have to use `flat` attribute
for ax in axes.flat:
    # Remove all xticks and yticks...
    ax.set(xticks=[2], yticks=[2])  
plt.show()
```
