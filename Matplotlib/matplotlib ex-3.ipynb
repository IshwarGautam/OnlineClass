```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(6)
x = np.arange(3)
y = np.random.randn(3)

fig, axes = plt.subplots(ncols=2, figsize=plt.figaspect(0.5))
vert_bars = axes[0].bar(x, y, color='lightblue', align='center')
hort_bars = axes[1].barh(x, y, color='lightblue', align='center')

axes[0].axhline(0, color='grey', linewidth=3)
axes[1].axvline(0, color='grey', linewidth=3)

plt.show()
```

```python
fig, ax = plt.subplots()
vert_bars = ax.bar(x, y, color='green', align='center')

for bar, height in zip(vert_bars, y):
    if height < 0:
        bar.set(color='red', linewidth=3)
        
plt.show()
```

```python
x = np.linspace(0, 20 ,100) #generate 100 number in array from 0 to 20
y1 = 2 * x + 1
y2 = 3 * x + 1.2
y_mean = 0.5 * x * np.cos(2*x) + 2.5 * x +1.1

fig, ax = plt.subplots()

ax.fill_between(x, y1, y2, color='black')
ax.plot(x, y_mean, color='white')

plt.show()
```
