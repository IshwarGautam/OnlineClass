```python
import matplotlib.pyplot as plt
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=plt.figaspect(0.3))

ax1.plot([-5,0,5,6,10], [1,5,0,3,6])
ax2.scatter([-5,0,5,6,10], [1,5,0,3,6])

plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=plt.figaspect(0.3))

ax1.plot([-5,0,5,6,10], [1,5,0,3,6])
ax2.scatter([-5,0,5,6,10], [1,5,0,3,6])

ax1.margins(x= 0.0, y=0.5)
ax2.margins(0.5)
ax1.axis('equal')
ax2.axis('tight')

plt.show()
```

```python
fig, axes = plt.subplots(nrows=3)

for ax in axes:
    ax.plot([-5,0,5,6,10], [1,5,0,3,6])
    
axes[0].set_title('Normal Autoscaling', y=0.5, x=0.6)

axes[1].set_title('ax.axis("tight")', y=0.5, x=0.6)
axes[1].axis("tight")

axes[2].set_title('ax.axis("equal")', y=0.5, x=0.8)
axes[2].axis("equal")

plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=plt.figaspect(0.5))
ax1.plot([-5,0,5,6,10], [1,5,0,3,6])
ax2.scatter([-5,0,5,6,10], [1,5,0,3,6])
ax1.set_ylim(bottom=-5)
ax2.set_xlim(right=20)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4], [10, 20, 30, 40], label='Nepalgunj')
ax.plot([1, 2, 3, 4], [31, 22, 10, 4], label='Butwal')
ax.set(ylabel='Temperature (Deg C)', xlabel='Time', title='A tale of two cities')
ax.legend()
plt.show()
```

```python
import numpy as np
plt.style.use('classic')

t = np.linspace(0, 2 * np.pi, 100)
x1, y1 = np.cos(t), np.sin(t)
x2, y2 = 2 * x1, 2 * y1

fig, ax = plt.subplots(figsize=(10, 10))
ax.plot(x1, y1, label='Inner', color='purple', lw=3)
ax.plot(x2, y2, label='Outer', color='red', lw=3)
ax.margins(0.1)
ax.axis('equal')
ax.legend()
plt.show()
```
