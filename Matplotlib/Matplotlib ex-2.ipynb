```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
t = np.arange(0, 1, 0.2)
plt.plot( t,t, t,t**2, t,t**3, marker='o') #plot (t,t),(t,t**2),(t,t**3)
plt.show()
```

```python
t = np.arange(0, 5, 0.2)
plt.plot(t, t, '-', t, t**2, '--', t, t**3, '*', t, -t, 'o')
plt.show()
```

```python
t = np.arange(0., 5., 0.2)
# red dashes, green squares and blue triangles
plt.plot(t, t, 'r--', t, t**2, 'gs', t, t**3, 'b^')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar([1, 2, 3, 4], [5,10,15,20], ls='dashed', ec='y', lw=5)
plt.show()
```

```python
#cosine wave
t = np.arange(1,2,0.01)
a = 1 + np.cos(2*np.pi*t)
plt.plot(t, a)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
z = np.random.random((5,5)) # 5 block in row and 5 block in column
ax1.imshow(z, interpolation='none', cmap='gray')
ax2.imshow(z, interpolation='none', cmap='coolwarm')
plt.show()
```

```python
plt.scatter([1,2,3,4],[1,3,5,7])
plt.title(r'$\sigma_i=1$', fontsize=15) #sigma
plt.show()
```

```python
from matplotlib.rcsetup import cycler
plt.rc('axes', prop_cycle=cycler('color', 'bgr')+cycler('lw', [1,3,5])+cycler('ls', ['-', '--', ':']))

t = np.arange(0., 5., 0.2)
plt.plot(t, t)
plt.plot(t, t**2)
plt.plot(t, t**3)
plt.show()
```

```python
plt.rc('axes', prop_cycle=cycler('color', ['b', 'g', 'r', 'y']) +
                          cycler('hatch', ['x', 'xx-', '+O.', '*']))
x = np.array([0.7, 0.3, 0.385, 0.6153, 0.7, np.nan, 0.6, 0.4, 0.2, 0.5, 0.8, 0.6])
y = [0.65, 0.65, 0.1, 0.1, 0.65, np.nan, 0, 0, -5, -6, -5, 0]
plt.fill(x+1, y)
plt.fill(x+2, y)
plt.fill(x+3, y)
plt.fill(x+4, y)
plt.show()
```
